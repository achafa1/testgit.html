```python
import matplotlib.pyplot as plt
import numpy as np

# Define the factors to compare
factors = ['Features', 'Brand Reputation', 'Pricing', 'Target Market']

# Define the brands and their scores for each factor
brands = ['Desley Pluggage', 'Rimowa', 'Away', 'Samsonite']
scores = {
    'Desley Pluggage': [9, 8, 7, 8],  # Example scores (out of 10)
    'Rimowa': [7, 9, 10, 7],
    'Away': [8, 7, 8, 9],
    'Samsonite': [7, 8, 7, 8]
}

# Convert scores to numpy arrays
scores_array = np.array([scores[brand] for brand in brands])

# Compute the angle for each factor
angles = np.linspace(0, 2 * np.pi, len(factors), endpoint=False).tolist()

# Make the plot close circular
scores_array = np.concatenate((scores_array, scores_array[:,[0]]), axis=1)
angles += angles[:1]

# Plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
for i, brand in enumerate(brands):
    ax.plot(angles, scores_array[i], label=brand, linewidth=1, linestyle='solid')
ax.fill(angles, np.mean(scores_array, axis=0), 'gray', alpha=0.1)

# Add labels
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(factors)
ax.set_title('Comparison of Desley Pluggage with Competitors')

# Add a legend
ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Show the plot
plt.show()
```
