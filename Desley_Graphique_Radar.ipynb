```python
import matplotlib.pyplot as plt
import numpy as np

# DÃ©finir les facteurs Ã  comparer
facteurs = ['FonctionnalitÃ©s', 'RÃ©putation de la Marque', 'Prix', 'MarchÃ© Cible']

# DÃ©finir les marques et leurs scores pour chaque facteur
marques = ['Desley Pluggage', 'Rimowa', 'Away', 'Samsonite']
scores = {
    'Desley Pluggage': [9, 8, 7, 8],  # Exemple de scores (sur 10)
    'Rimowa': [7, 9, 10, 7],
    'Away': [8, 7, 8, 9],
    'Samsonite': [7, 8, 7, 8]
}

# Convertir les scores en tableaux numpy
scores_tableau = np.array([scores[marque] for marque in marques])

# Calculer l'angle pour chaque facteur
angles = np.linspace(0, 2 * np.pi, len(facteurs), endpoint=False).tolist()

# Faire le tracÃ© en forme de radar
scores_tableau = np.concatenate((scores_tableau, scores_tableau[:,[0]]), axis=1)
angles += angles[:1]

# TracÃ©
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
for i, marque in enumerate(marques):
    ax.plot(angles, scores_tableau[i], label=marque, linewidth=1, linestyle='solid')
ax.fill(angles, np.mean(scores_tableau, axis=0), 'gray', alpha=0.1)

# Ajouter les Ã©tiquettes
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(facteurs)
ax.set_title('Graphique Radar de Comparaison de Desley Pluggage avec ses Concurrents', 'color:blue')

# Ajouter une lÃ©gende
ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Afficher le tracÃ©
plt.show()
```
